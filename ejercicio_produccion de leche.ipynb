```python
import numpy as np

# Datos
a침os = np.array(list(range(2005, 2020)))
produccion_leche = np.array([2.3, 4.5, 9, 7.8, 9.12, 10, 10.22, 10.95, 12.3, 12.5, 12.6, 12.77, 13.5, 16.8, 17.2])

# Calcular las sumas necesarias
n = len(a침os)
sum_x = np.sum(a침os)
sum_y = np.sum(produccion_leche)
sum_xy = np.sum(a침os * produccion_leche)
sum_x_squared = np.sum(a침os**2)

# Calcular la pendiente (m) y la intersecci칩n con el eje y (b)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - m * sum_x) / n

# Imprimir la pendiente y la intersecci칩n con el eje y
print("Pendiente (m):", m)
print("Intersecci칩n con el eje y (b):", b)

# Calcular la ecuaci칩n de la l칤nea recta ajustada
print("Ecuaci칩n de la l칤nea recta ajustada: y =", m, "x +", b)
```
